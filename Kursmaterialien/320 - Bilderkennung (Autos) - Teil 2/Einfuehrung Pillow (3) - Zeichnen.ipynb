```python
from PIL import Image

img = Image.open("bild1.jpg")
img = img.resize((round(img.size[0] / 10), round(img.size[1] / 10)))
```

```python
img
```

```python
from PIL import ImageDraw

draw = ImageDraw.Draw(img)

#Liste der Punkte
points = [
    (0, 0),
    (100, 100)
]

#Übergeben der Liste, Farbe und Breite der Linie
draw.line(points, "red", 5)
```

```python
img
```

```python
points = [
    (100, 100),
    (100, 200)
]
draw.line(points, "green", 5)
```

```python
#2 Punkte die das Rechteck definieren angeben -> outline, damit nicht ausgefüllt

draw.rectangle([
    (50, 50),
    (150, 150)
], outline="yellow")
```
